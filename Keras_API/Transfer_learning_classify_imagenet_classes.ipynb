```
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
tf.version.VERSION
```

```
resnet50 = tf.keras.applications.resnet50.ResNet50(weights='imagenet');
```

```
img = tf.keras.preprocessing.image.load_img('elephant.jpeg', target_size=(224,224))

plt.imshow(img)
plt.show()
```

```
img_arr = tf.keras.preprocessing.image.img_to_array(img)
img_arr.shape
```

```
img_arr_ = np.expand_dims(img_arr, axis=0)
img_arr_.shape
```

```
x = tf.keras.applications.resnet50.preprocess_input(img_arr_)
preds = resnet50.predict(x)
print('Predicted: ', tf.keras.applications.resnet50.decode_predictions(preds, top=3)[0])
```
